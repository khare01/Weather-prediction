```python
import pandas as pd
data = pd.read_csv('weather.csv')
print(data.head())
print(data.info())
print(data.describe())
```

```python
import matplotlib.pyplot as plt

# Plot histogram for temperature
plt.hist(data['temperature_celsius'], bins=20, color='skyblue')
plt.xlabel('Temperature (C)')
plt.ylabel('Frequency')
plt.title('Temperature Distribution')
plt.show()

# Plot histogram for precipitation
plt.hist(data['precip_mm'], bins=20, color='lightgreen')
plt.xlabel('Precipitation (mm)')
plt.ylabel('Frequency')
plt.title('Precipitation Distribution')
plt.show()
```

```python
import seaborn as sns

# Select relevant columns for correlation analysis
correlation_cols = ['temperature_celsius', 'precip_mm', 'wind_kph', 'humidity', 'air_quality_PM2.5']

# Calculate the correlation matrix
correlation_matrix = data[correlation_cols].corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
data['last_updated'] = pd.to_datetime(data['last_updated'])
data.set_index('last_updated', inplace=True)

# Plot line graph for temperature changes over time
plt.figure(figsize=(12, 6))
plt.plot(data['temperature_celsius'])
plt.xlabel('Time')
plt.ylabel('Temperature (C)')
plt.title('Temperature Changes over Time')
plt.show()
```

```python
# Scatter plot of latitude vs. longitude to visualize geographical locations
plt.scatter(data['longitude'], data['latitude'], alpha=0.5, color='blue')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Geographical Distribution of Locations')
plt.show()
```
